You are part of the IT Team of a huge company, operating on the US stock market for many years.

In exchange for advice, you are more than happy to help him whenever he needs assistance with this new challenge of his - coding in Python.

With Mr. Peterson leading the way, your team now has to analyze the performance of three market indices from January 3rd, 2000 to May 21st, 2021. The indices and their respective ticker symbols are:

- Standard and Poor’s – S&P (‘^GSPC’)
- Nasdaq Composite (‘^IXIC’)
- Dow Jones Industrial Average (‘^DJI’)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
market_indices = pd.read_csv('365DSc-Indices-2000-2021.csv', index_col='Date')
market_indices
```

## Normalize the data with a starting point equal to 100

```python
normalized_data = (market_indices / market_indices.iloc[0] * 100)
f"{normalized_data.loc['2000-01-05','^DJI']:.6f}"
```

## Analyze

```python
normalized_data.plot(figsize = (15, 6))
plt.show()
```

Conclusion: IT companies’ stock return never really managed to top the US stock market average since the dot-com bubble in the late 1990s.

## Calculate annualized simple return

Mr. Peterson says that the last important part of your team’s analysis will be inspecting whether the average of companies listed under the S&P, Nasdaq and Dow Jones indices have provided investors with a positive or negative return from 2000 to early 2021.

To answer this question, Mr. Peterson need to run in Python in order to obtain the *annualized simple returns*

```python
(market_indices / market_indices.shift(1) - 1).mean() * 250
```

Conclusion: Nasdaq surpasses the Dow Jones and S&P indices
